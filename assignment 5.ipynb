```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

N_households = 500
vehicles_per_household = [0, 1, 2, 3, 4, 5]

# Assuming 'vehicle_data' should contain the number of vehicles for each household
# Replace this with your actual data source or generation method
# Here's an example of randomly generating vehicle data:
vehicle_data = np.random.choice(vehicles_per_household, size=N_households)

# Now you can calculate the pmf values:
pmf_values = [np.sum(vehicle_data == vehicles) / N_households for vehicles in vehicles_per_household]
```

```python
cdf_values = np.cumsum(pmf_values)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(vehicles_per_household, pmf_values, color='skyblue', edgecolor='black')
plt.title('Probability Mass Function (PMF) of Vehicle Ownership per Household')
plt.xlabel('Number of Vehicles')
plt.ylabel('Probability')
plt.xticks(vehicles_per_household)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.step(vehicles_per_household, cdf_values, where='post', color='orange', marker='o')
plt.title('Cumulative Distribution Function (CDF) of Vehicle Ownership per Household')
plt.xlabel('Number of Vehicles')
plt.ylabel('Cumulative Probability')
plt.xticks(vehicles_per_household)
plt.grid(True)
plt.show()
```

```python
N_gadgets = 100
p_defective = 0.05
```

```python
k_exactly_10 = 10
probability_exactly_10 = binom.pmf(k_exactly_10, N_gadgets, p_defective)
print(f"Probability that exactly 10 gadgets are defective: {probability_exactly_10:.4f}")
```

```python
k_at_least_15 = 15
probability_at_least_15 = 1 - binom.cdf(k_at_least_15 - 1, N_gadgets, p_defective)
print(f"Probability that at least 15 gadgets are defective: {probability_at_least_15:.4f}")
```

```python
expected_defective = N_gadgets * p_defective
print(f"Expected number of defective gadgets: {expected_defective:.2f}")
```
